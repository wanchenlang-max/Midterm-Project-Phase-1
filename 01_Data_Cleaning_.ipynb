```python

import os
import zipfile

project_folder = "/content/replication_project"
raw_folder = project_folder + "/data/raw"

os.makedirs(raw_folder, exist_ok=True)

print("Folder created successfully")


zip_url = "https://davidcard.berkeley.edu/data_sets/njmin.zip"
zip_path = raw_folder + "/njmin.zip"

!wget -O "$zip_path" "$zip_url"

print("Data download complete")

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(raw_folder)

print("Decompression complete")

print("\n raw Folder contents：")
for file in os.listdir(raw_folder):
    print(" -", file)
```

```python
import pandas as pd

file_path = "/content/replication_project/data/raw/public.dat"


df = pd.read_fwf(file_path, header=None)

print("Data Shape:", df.shape)


df.columns = [f"var_{i}" for i in range(df.shape[1])]

df.head()
```
