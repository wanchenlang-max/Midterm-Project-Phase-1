```python
import re
from pathlib import Path

sas_path = Path("/content/Midterm-Project-Phase-1/data/raw/check.sas")
sas_text = sas_path.read_text(errors="ignore")

no_strings = re.sub(r"'([^']|'')*'", "''", sas_text)
no_strings = re.sub(r'"([^"]|"")*"', '""', no_strings)

no_strings = re.sub(r"(?is)\bcomment\b.*?;", "", no_strings)
no_strings = re.sub(r"(?is)\bdata\b.*?;", "", no_strings)
m = re.search(r"(?is)\binput\b\s+(.*?);", no_strings)
assert m, "NOT FOUND INPUT ... ）"

block = m.group(1)


block = re.sub(r"@\d+", " ", block)
block = re.sub(r":[A-Za-z0-9.]+", " ", block)
block = re.sub(r"\$[A-Za-z0-9.]+", " ", block)
block = re.sub(r"[;]", " ", block)

names = [x for x in re.split(r"\s+", block.strip()) if x]

print("Number of variables:", len(names))
print("The top 20:", names[:20])
print("The last 20:", names[-20:])
```

```python
import numpy as np
import pandas as pd

id_cols = ['SHEET']
date_cols = []
df['EMPFT'] = pd.to_numeric(df['EMPFT'], errors='coerce')
df['EMPPT'] = pd.to_numeric(df['EMPPT'], errors='coerce')
df['EMPFT2'] = pd.to_numeric(df['EMPFT2'], errors='coerce')
df['EMPPT2'] = pd.to_numeric(df['EMPPT2'], errors='coerce')


df['fte1'] = df['EMPFT'] + 0.5*df['EMPPT']
df['fte2'] = df['EMPFT2'] + 0.5*df['EMPPT2']

long = pd.DataFrame({
    'SHEET': pd.concat([df['SHEET'], df['SHEET']], ignore_index=True),
    'treat': pd.concat([df['treat'], df['treat']], ignore_index=True),
    'fte': pd.concat([df['fte1'], df['fte2']], ignore_index=True),
    'post': np.repeat([0,1], repeats=len(df))
})
long['did'] = long['treat'] * long['post']
long.head()
```

```python
long.groupby(['treat','post'])['fte'].mean()
```

```python
import statsmodels.formula.api as smf

long_filtered = long.dropna(subset=['fte', 'treat', 'post', 'did'])

model = smf.ols("fte ~ treat + post + did", data=long_filtered)

res_ols = model.fit()

filtered_groups = long_filtered["SHEET"]

res = res_ols.get_robustcov_results(cov_type="cluster", groups=filtered_groups)
print(res.summary())
```

```python
import numpy as np
import pandas as pd

num_cols = [
    'EMPFT','EMPPT','NCALLS','WAGE_ST',
    'EMPFT2','EMPPT2','NCALLS2','WAGE_ST2'
]

for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')  # '.' -> NaN

df[num_cols].isna().mean().sort_values(ascending=False)
```

```python
df['fte1'] = df['EMPFT']  + 0.5 * df['EMPPT']
df['fte2'] = df['EMPFT2'] + 0.5 * df['EMPPT2']

df[['EMPFT','EMPPT','fte1','EMPFT2','EMPPT2','fte2']].head()
```

```python
id_col = 'SHEET'

df_long = pd.DataFrame({
    id_col: pd.concat([df[id_col], df[id_col]], ignore_index=True),
    'treat': pd.concat([df['treat'], df['treat']], ignore_index=True),
    'post':  pd.Series([0]*len(df) + [1]*len(df)),
    'fte':   pd.concat([df['fte1'], df['fte2']], ignore_index=True),
    'wage':  pd.concat([df['WAGE_ST'], df['WAGE_ST2']], ignore_index=True),
    'ncalls':pd.concat([df['NCALLS'], df['NCALLS2']], ignore_index=True),
})
df_long.head()
```

```python
df_long.groupby(['treat','post'])['fte'].mean()
```

```python
table_fte = (
    df_long
    .groupby(["treat", "post"])["fte"]
    .agg(["mean", "std", "count"])
    .reset_index()
)
table_fte
```

```python
pivot_fte = df_long.pivot_table(
    index="treat", columns="post", values="fte",
    aggfunc=["mean", "std", "count"]
)
pivot_fte
```

```python
means = df_long.groupby(['treat','post'])['fte'].mean().unstack()
means.columns = ['pre','post']
means['diff_post-pre'] = means['post'] - means['pre']
means
```

```python
did_manual = (means.loc[1,'diff_post-pre'] - means.loc[0,'diff_post-pre'])
did_manual
```

```python
import statsmodels.formula.api as smf

reg_df = df_long.dropna(subset=['fte', 'treat', 'post', id_col]).copy()

m = smf.ols('fte ~ treat * post', data=reg_df)
res = m.fit(cov_type='cluster', cov_kwds={'groups': reg_df[id_col]})

print(res.summary())
```

```python
print("Manual DID (from means) should match interaction coef in regression.")
```

```python
print("Manual DID:", did_manual)
print("Regression DID (treat:post):", res.params['treat:post'])
```

```python
out = pd.DataFrame({
    'coef': res.params,
    'se_cluster': res.bse,
    't': res.tvalues,
    'p': res.pvalues
})
out.loc[['Intercept','treat','post','treat:post']]
```
